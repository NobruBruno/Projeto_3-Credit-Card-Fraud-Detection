```python
# import bibliotecas 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,mean_squared_error, r2_score
```

```python
# import arquivo
df_credit = pd.read_csv('creditcard.csv')
```

```python
# visualizar o arquivo
df_credit.head()
```

```python
# verificando os tipos
df_credit.info()
```

```python
# valores nulos 
df_credit.isnull().sum()
```

```python
# Dados não fraudes
df_N_fraude = df_credit.Amount[df_credit.Class == 0]
df_N_fraude.describe()
```

```python
# Dados fraudes
df_fraude = df_credit.Amount[df_credit.Class == 1]
df_fraude.describe()
```

```python
# Tipos de dados de fraude 
df_fraude = df_credit[df_credit.Class == 1]
df_fraude
```

```python
# tipos de dados não fraude 
df_NAO_fraude = df_credit[df_credit.Class == 0]
df_NAO_fraude
```

```python
# Quantidade = a de fraude e Não fraude
df_NAO_fraude = df_NAO_fraude.sample(n=492)
df_NAO_fraude
```

```python
# concatenar dos datafremes
df = pd.concat([df_NAO_fraude, df_fraude], axis=0)
df
```

```python
# ajustar o index
df.reset_index(inplace=True)
df
```

```python
# Retirar dados de validação do modelo
df_val_N_fraude = df.head(5)
df_val_N_fraude
```

```python
df_val_fraude = df.tail(5)
df_val_fraude
```

```python
# retirar as 5 primeiras  5 ultimas linhas 
df = df.iloc[5:]
df = df.iloc[:-5]
df
```

```python
# Concatenar e ajustar o index do modelo de validaçaõ
df_val_tatol = pd.concat([df_val_N_fraude, df_val_fraude])
df_val_tatol.reset_index(inplace=True)
df_val_tatol
```

```python
# Retirar informações para bater depois com a Previsão e os valores Reais 
df_val_tatol_real = df_val_tatol.Class
df_val_tatol = df_val_tatol.drop(['level_0','index','Time','Class'], axis=1)
df_val_tatol
```

```python
# verificando a distribuição 
df.Class.value_counts()
```

```python
# Separando X e Y
X = df.drop(['index','Time','Class'], axis=1)
Y =  df['Class']
```

```python
print (X,Y)
```

```python
# Dados de treino e Dados de teste
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.30, random_state=33, stratify=Y)
```

```python
# treinamento 
Lr = LogisticRegression(max_iter=1000)
Lr.fit(X_train, Y_train)
pred = Lr.predict(X_test)
acc = accuracy_score(Y_test, pred)
```

```python
print (f'Acuracia:{acc*100:.2f}')
```

```python
# Validação 
pred = Lr.predict(df_val_tatol)
df = pd.DataFrame({'Real':df_val_tatol_real,'Previsão':pred})
```
